### Presentamos la cantidad de nacidos por Sexos

```python
#Importamos show para presentar las graficas y el output_notebook para visualizarlas las celdas del jupyter
from bokeh.io import show, output_notebook
#El ColumnDataSource para compartir datos entre múltiples gráficos y widgets 
from bokeh.models import ColumnDataSource, HoverTool
#Se encargan de armar gráficos con ejes, cuadrículas y herramientas predeterminadas.
from bokeh.plotting import figure
#"Spectral6" proporciona una lista de seis elementos de cadenas hexagonales RGB (A) del mapa de colores "Spectral" de Brewer.
from bokeh.palettes import Spectral6
#Asigna colores a elementos categóricos específicos
from bokeh.transform import factor_cmap


#Presentamos la cantidad de mujeres y hombres nacidos en el 2014
sexo = list(datos["sexo"])#Creamos una lista con que contiene todos los sexos
sexIndex = list(set(sexo))#Asigamos la lista a otra variable sexIndex

#Contamos el numero de cada sexo por categorias con la funcion .count() y almacenamos en variable d1 y d2
d1 = sexo.count(sexIndex[0])
d2 = sexo.count(sexIndex[1])

pos_x = sexIndex#Datos que van en el eje x que son los dos tipos de sexo
pos_y = [d1,d2]#Datos que van en el eje y que son las cantidades por sexo contadas

#ColumnDataSource es el objeto donde se almacenan los datos de un gráfico Bokeh. 

#Colocamos todos los datos, en un ColumnDataSource. 
source = ColumnDataSource(data = dict(pos_x=pos_x, pos_y=pos_y))


#llamamos al metodo para presentar en celdas de jupyter solo una vez
output_notebook()

#Figure principal clase de plotting, esto crea un Figure modelo, que incluye métodos para agregar diferentes tipos de glifos a un gráfico.
graf_Sexos = figure(x_range=pos_x, plot_height=350, toolbar_location=None, title="Cantidades por Sexos")


#Utilizamo el vbar y establecemos la coordenada X y el top que es el alto de las barras
#el nombre de la columna Source se pasa a figure
#Usamos los métodos de glifos .vbar para agregar renderizadores para nuestros datos 
graf_Sexos.vbar(x='pos_x',  top='pos_y', width=0.4, source=source, legend="pos_y",
       line_color='black', fill_color = factor_cmap('pos_x', palette = Spectral6, factors=pos_x))


graf_Sexos.xgrid.grid_line_color = None     #Quitamos la linea vertical
graf_Sexos.y_range.start = 0                #Comenzamos el eje y en 0
graf_Sexos.y_range.end = 1500               #Terminamos el eje y en 1500
graf_Sexos.legend.orientation = "horizontal"#Indicadores con cantidades
graf_Sexos.legend.location = "top_right"     #Localizacion del indicador

#Funcion para presentar
show(graf_Sexos)
```

En esta gráfica se presenta la contabilización de nacimientos por sexo donde las dos cantidades dan 2485 que es el número de registros de nuestro csv, por lo cual esta correcta esa cantidad, además se puede observar que en el 2014 el sexo con más nacimientos son los hombres

## Gráfica con datos de la provincia de la Sierra con más nacimientos inscritos, detallando por asistentes del parto

Presentamos los datos de la provincia Pichincha que es la que tiene más nacimientos inscritos con el numero de asistentes que intervinieron en los partos con categorías de:

            - Médico/a
            - Obstetriz/Obstetra
            - Partero/a no calilficado/a

```python
#Asignamos los tipos de asistentes en la provincia 1 Pichincha filtrada
asisP1 = prov1[prov1['asis_por']=='M�dico/a']
asisP2 = prov1[prov1['asis_por']=='Obstetriz/Obstetra']
asisP3 =prov1[prov1['asis_por']=='Partero/a no calilficado/a']
```

```python
#Creamos un arreglo de la provincia que vamos a graficar
provincias = ['Pichincha']

#A continuación creamos un arreglo con los asistentes de parto que vamos a representar
asistentes = ['Medico/a','Obstetriz/Obstetra','Partero/a']

#Asiganamos los valores contados de cada provincia en sus respectivos asistentes contados
data = {'provincias':provincias,
         'Medico/a':asisP1.count(),
         'Obstetriz/Obstetra':asisP2.count(),
         'Partero/a':asisP3.count()
    
}
```

```python
#librerias para la grafica
from bokeh.transform import dodge       #Se ua para obtener un solo valor
from bokeh.core.properties import value #Tomar un valor espececifico de una variable

#Colocamos todo el data, en un ColumnDataSource llamada fuente 
fuente = ColumnDataSource(data)

#Establecemos de los valores a graficar y caracteriscas de la grafica
#rango de x = provincia, y= hasta 350 
grafica_barra = figure(x_range=provincias, y_range=(0, 350), plot_height=350, title="Asistentes en Partos de Pichincha")

#Tipo de glifo .vbar, establecemos sus caracteristicas de la barra
#Posicionamos las barras para que no se monten
#En el dodge asignamos = un nombre de campo para configurar DataSpec, el desplazamiento fijo de la columna, rango calcular las coordenadas
grafica_barra.vbar(x = dodge('provincias', -0.25, range=grafica_barra.x_range), top='Medico/a', width=0.2, source=fuente,
       color="#FC3E3E", legend = value("Médico/a"))

grafica_barra.vbar(x = dodge('provincias',  0.0,  range=grafica_barra.x_range), top='Obstetriz/Obstetra', width=0.2, source=fuente,
       color="#02BEBB", legend = value("Obstetriz/Obstetra"))

grafica_barra.vbar(x = dodge('provincias',  0.25, range=grafica_barra.x_range), top='Partero/a', width=0.2, source=fuente,
       color="#289C2F", legend = value("Partero/a no calilficado/a"))


grafica_barra.x_range.range_padding = 0       #Ancho de la barra
grafica_barra.xgrid.grid_line_color = None    #Quitamos linea vertical
grafica_barra.legend.location = "top_left"    #Posicion para indicador de datos
grafica_barra.legend.orientation = "vertical"   #Colocamos orientacion horizontal
show(grafica_barra)                               #presentamos
```

#### La mayoría de partos en esta provincia de Pichincha del sector Sierra son atendidos por un Médico con 249, Obstetriz con 35 y Pateros con 1

### Gráfica circular de datos por lugar de nacimiento de los bebes

```python
#librerias para la grafica
from math import pi                    #Para calcular el tamaño del circulo
from bokeh.palettes import Viridis     #Colores para las divisones
from bokeh.transform import cumsum     #Devuelve un marco de datos o una serie del mismo tamaño que contiene la suma acumulada

#Almacenamos los datos a represantar en X
x = {
    'Establecimiento de la Junta de Beneficencia': 239,
    'Hospital, Clínica o Consultorio Privado': 165,
    'Establecimiento del Ministerio de Salud': 195,
    'Establecimiento del IESS': 24,
    'Otro Establecimiento Público':3,
    'Casa':2
}


data = pd.Series(x).reset_index(name='value').rename(columns={'index':'cant'})
#Para el tamaño de los ángulos de cada categoria
data['angle'] = data['value']/data['value'].sum() * 2*pi
#Colores del circulo con rango del conjunto X, toma la cantidad de colores que tenga x
data['color'] = Viridis[len(x)]

#Creamos objeto graf_Lugar para graficar -  establecemos características
graf_Lugar = figure(plot_width=700, plot_height=350, title="Lugares de Nacimientos", toolbar_location=None,
           tools="hover", tooltips="@cant: @value", x_range=(-0.5, 1.5))

#Tipo de glifo .wedge, establecemos sus caracteristicas del pastel
graf_Lugar.wedge(x=0, y=1, radius=0.4, start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="black", fill_color='color', legend='cant', source=data)


graf_Lugar.axis.axis_label = None         #Quitamos cambios del eje x
graf_Lugar.axis.visible = False           #Ocultamos el eje X y Y
graf_Lugar.grid.grid_line_color = None    #Ocultamos las lineas horizontales y verticales de la grafica

show(graf_Lugar)                          #Presentamos la grafica
```

### Gráfica por tipo de Inscripciones de los nacimientos de la Provincia del Guayas, sector urbano y de madres Solteras

```python
#Libreria para la gráfica
from bokeh.models import FactorRange #Un rango de valores para una dimensión categórica.

#Categorias a graficar
inscripciones = [
    ("Inscripción Normal", ""),
    ("Inscripción Tardía", "") 

]

#Creamos el objeto graf_Inscrip tipo Figure para asignarle caracteristicas de la grafica
#El Factorange ayuda a obtener el rango de la variable inscripciones
graf_Inscrip = figure(x_range = FactorRange(*inscripciones), plot_height=250, toolbar_location=None, tools="")

#Asignamos los valores obtenidos a cada respectiva Inscripción
ValInscripciones = [149, 47]

#Como es una gráfica de barras utilizamos el glifo .vbar y establecemos las corrdenadas x y el alto de las barras en top
graf_Inscrip.vbar(x = inscripciones, color=["#0FFF83","#F61F46"], top = ValInscripciones, width=0.9, alpha=0.5)

#Para la grafica de la linea utilizamos el .line, le asignamos coordenadas de X y Y y más caracteristicas 
#Colocamos encima de que barras va a ir la linea
graf_Inscrip.line(x=["Inscripción Normal", "Inscripción Tardía"], y=[149,47], color="red", line_width = 3.5)


graf_Inscrip.y_range.start = 0                 #Establecemos el comiezo de los intervalos para el eje y 
graf_Inscrip.x_range.range_padding = 0.2       #Establecemos el tamaño del relleno de la barra
graf_Inscrip.xgrid.grid_line_color = None      #Quitamos la linea vertical 

show(graf_Inscrip)                             #Presentamos la grafica
```
